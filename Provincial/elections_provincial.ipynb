```python
## select the needed columns

results_clean_df = results_df[['ElectoralDistrictNameEnglish', 'VotingPlaceAddressOrLocation',
                                            'NameOfCandidates', 'party', 'AcceptedBallotCount']]

results_clean_df = results_clean_df.rename(columns={
    
                                                    'ElectoralDistrictNameEnglish' : 'district',
                                                    'VotingPlaceAddressOrLocation' : 'Location',
                                                    'NameOfCandidates' : 'candidate' , 
                                                    'AcceptedBallotCount' : 'votes'
                                                    
                                                    })
results_clean_df.head()
```

```python
## pivot the DF to have a view on candidates

results_pivot = results_clean_df.pivot_table(
                                                                    index=['district', 'Location'],
                                                                    columns='party',
                                                                    values='votes',
                                                                    aggfunc='sum',  
                                                                    fill_value=0
                                                                )

results_pivot.head()
```

```python
## create a column Total

results_pivot['Total']= results_pivot.sum(axis=1)
results_pivot.head()
```

```python
## create a column Others that will aggregate all the other candidates

results_pivot['Others'] = results_pivot['Total'] - results_pivot['PCP'] - results_pivot['LIB']
results_pivot
```

```python
## Add calculation to the DF 

### calculate the difference in votes

results_pivot['difference'] = results_pivot['LIB'] - results_pivot['PCP']

### identify the winner

results_pivot['winner'] = np.where(
                                                results_pivot['difference'] > 0,
                                                'LIB', 'PCP'
                                            )

### calculate the difference on the total voters

results_pivot['Diff %'] = np.where(
                                                results_pivot['difference'] < 0,
                                                -results_pivot['difference'] / results_pivot['Total'],
                                                results_pivot['difference'] / results_pivot['Total']  
                                            )
### calculate the difference on the remaining voters (others)

results_pivot['Diff % others'] = np.where(
                                                        results_pivot['difference'] < 0,
                                                        -results_pivot['difference'] / results_pivot['Others'],
                                                        results_pivot['difference'] / results_pivot['Others']  
                                                    )

results_pivot = results_pivot.reset_index()
results_pivot.head()
```

## Retrieve and add the coordinates 

### use geopify api to get the coordinates via the reverse mode

```python
# create list of vote_places

## extract unique value of vote places
vote_places = results_clean_df['Location'].unique()

## create a list
vote_places_list = vote_places.tolist()

print(vote_places_list[0:5])
print(f'.............................................')
print(f'number of vote places : {len(vote_places_list)}')
```

```python
# API key geopify api

api_key = "8e0f597430e942f0ac38b55cf85ff32b"
```

```python
# FOR LOOP can take 1m 50s for 75 places

# List to hold dictionaries for each place
places_coordinates = []  


# loop through the places in the list and retrive the ccordinates

for place in vote_places_list:

    ## complete the places to specify they are in oakville, ON

    

    full_place = f'{place}; Oakville; Ontario; Canada'
    
    try:
        encoded_place = quote(full_place)
        url = f"https://api.geoapify.com/v1/geocode/search?text={encoded_place}&apiKey={api_key}"
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()

        if data.get('features'):
            latitude = data['features'][0]['properties']['lat']
            longitude = data['features'][0]['properties']['lon']

            place_entry = {
                'Location': place,
                'full_Location' : full_place,
                'lon': longitude,
                'lat': latitude
            }
            places_coordinates.append(place_entry)

        else:
            place_entry = {
                'Location': place,
                'full_Location' : full_place,
                'lon': 'none',
                'lat': 'none'
            }
            places_coordinates.append(place_entry)
            print(f"No coordinates found for {place}")

        time.sleep(0.5)

    except requests.RequestException as e:
        print(f"Request failed for {place}: {e}")

print(places_coordinates)
```

```python
## create a DF with coordinates

coordinates_df = pd.DataFrame(places_coordinates)

coordinates_df
```
